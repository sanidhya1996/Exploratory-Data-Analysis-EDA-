## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Transforming the 'Reviews' column from Object type to float type.
Psd['Reviews'] = Psd['Reviews'].astype(float)


# Removing '$' in the 'Price' column and transforming into Float Datatype.
Psd['Price'] = Psd['Price'].apply(lambda x: str(x).replace('$','') if '$' in str(x) else str(x))
Psd['Price'] = Psd['Price'].apply(lambda x : float(x))


# Cleaning the Data of 'Installs' and transforming into numeric.
Psd['Installs'] = Psd['Installs'].map(lambda x: (x.rstrip('+')))
Psd['Installs'] = Psd['Installs'].map(lambda x: ''.join(x.split(',')))
Psd['Installs'] = pd.to_numeric(Psd['Installs'])
````

Answer Here:-I transformed the review column from object ot float type and install column to numeric and rmoved '$' from the price column and transformed it to float
